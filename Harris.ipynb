```python
# Cell 1: Import libraries and set up inline plotting
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cell 2: Load and process the image
image = cv2.imread('Screenshot 2025-04-04 at 3.33.02 PM.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Harris Corner Detection
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Dilate to mark corners
dst = cv2.dilate(dst, None)

# Threshold and mark corners on the image
image[dst > 0.01 * dst.max()] = [0, 0, 255]  # Red corners

# Display the result
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.axis('off')  # Hide axes
plt.show()

# Optional: Save the result
# cv2.imwrite('harris_corners.jpg', image)
```
